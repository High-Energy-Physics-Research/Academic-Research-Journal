A descrição hidrodinâmica emerge quando os graus de liberdade microscópicos
são efetivamente integrados, resultando em uma dinâmica governada por quantidades
coletivas. Nesse sentido, <mark class="mk">o campo hidrodinâmico já é um objeto coarse-grained</mark>,
e não uma variável fundamental da teoria subjacente.
</p>

<p>
<mark class="mk-def">Definição.</mark>
Chamamos de expansão longitudinal simples o regime no qual o divergente do
campo de velocidades satisfaz ∇<sub>μ</sub>u<sup>μ</sup> = 1/τ, como no cenário de Bjorken.
</p>

<p>
Esse resultado implica que a evolução temporal da densidade de energia
é diretamente controlada pela expansão do fluido, levando ao resfriamento
do sistema. <mark class="mk-insight">A hidrodinâmica codifica, portanto, um tipo de coarse-graining temporal</mark>,
no qual detalhes microscópicos são continuamente esquecidos à medida que o tempo
próprio evolui.
</p>

<p>
<mark class="mk-warn">Atenção.</mark>
Essa descrição é válida apenas enquanto hipóteses como equilíbrio local e
gradientes pequenos permanecem consistentes. Fora desse regime, correções
não hidrodinâmicas tornam-se relevantes.
</p>

<p>
Sob essa perspectiva, a hidrodinâmica relativística pode ser interpretada
não apenas como uma teoria efetiva, mas como uma linguagem natural para
descrever a irreversibilidade emergente em sistemas quânticos de muitos corpos.
</p>
